```python
def plot_network(network):
    plt.figure(figsize=(6, 6))
    pos=nx.spring_layout(network, k=0.15)
    nx.draw_networkx(network,pos,arrows=True,node_size=25, node_color='blue', with_labels=False)

plot_network(g)
plt.show()
```

```python
def plot_network(network):
    plt.figure(figsize=(20, 20))
    pos=nx.random_layout(network)
    nx.draw_networkx(network,pos,arrows=True,node_size=25, node_color='blue', with_labels=False)

plot_network(g)
plt.show()
```

```python
def plot_network(network):
    plt.figure(figsize=(6, 6))
    pos=nx.kamada_kawai_layout(network)
    nx.draw_networkx(network,pos,arrows=True,node_size=25, node_color='yellowgreen', with_labels=False)

plot_network(g)
plt.show()
```

```python
def plot_network(network):
    plt.figure(figsize=(6, 6))
    pos=nx.shell_layout(network)
    nx.draw_networkx(network,pos,arrows=True,node_size=25, node_color='pink', with_labels=False)

plot_network(g)
plt.show()

```

```python
def plot_network(network):
    plt.figure(figsize=(6, 6))
    pos=nx.spectral_layout(network)
    nx.draw_networkx(network,pos,arrows=True,node_size=25, node_color='turquoise', with_labels=False)

plot_network(g)
plt.show()
```
